```python
from typing import NamedTuple

class Order(NamedTuple):
    total_bill: float
    tip: float
    sex: str
    smoker: str
    day: str
    time: str
    size: int
```

```python
print(type(tips_dfm))
```

```python
print(tips_dfm.iterrows())
```

```python
next(tips_dfm.iterrows())
```

```python
orders = []

for i, row in tips_dfm.iterrows():
    order = Order(
        total_bill = row["total_bill"],
        tip = row["tip"],
        sex= row["sex"],
        smoker= row["smoker"],
        day= row["day"],
        time= row["time"],
        size = row["size"]
    )
    orders.append(order)

print(orders[0])
```

```python
orders_2 = [
    Order(
        total_bill = row["total_bill"],
        tip = row["tip"],
        sex= row["sex"],
        smoker= row["smoker"],
        day= row["day"],
        time= row["time"],
        size = row["size"]
    )
for _, row in tips_dfm.iterrows()]

print(orders_2[0])
```

```python
for order in orders[:10]:
    tip_prc = (order.tip / order.total_bill) * 100
    print(tip_prc)
```

```python
total_income = 0
total_tips = 0
total_guest = 0


for order in orders:
    total_income += order.total_bill
    total_tips += order.tip
    total_guest += order.size

avg_income = total_income / len(orders)

print(total_income)
print(avg_income)
print(total_tips)
print(total_guest)
```

```python
fruits = {
    "apple": 0,
    "banana": "green",
}

fruits["apple"] = fruits.get("apple", 0) + 4
print(fruits)
```

```python
income_day = {}
orders_day = {}
guest_day = {}

for order in orders:
    day = order.day
    income_day[day] = income_day.get(day, 0) + order.total_bill
    orders_day[day] = orders_day.get(day, 0) + 1
    guest_day[day] = guest_day.get(day, 0) + order.size



for day in income_day:
    n = orders_day[day]
    g = guest_day[day]
    r = income_day[day]
    print(f"{day}, {g}, {n}, {g/n:.1f} oсіб")

```

```python
big_orders = [order for order in orders if order.total_bill > 50]
print(len(big_orders))
print(big_orders)
```

```python
all_tip_percents = [(order.tip / order.total_bill) * 100 for order in orders]
print(all_tip_percents[0])
avg_tip = sum(all_tip_percents) / len(all_tip_percents)
print(f"середні {avg_tip:.2f}% min {min(all_tip_percents):.2f}% max {max(all_tip_percents):.2f}%")
```
