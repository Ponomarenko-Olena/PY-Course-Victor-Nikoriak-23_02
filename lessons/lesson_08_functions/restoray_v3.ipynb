```python
###
from typing import NamedTuple

class Order(NamedTuple):
    total_bill: float
    tip: float
    sex: str
    smoker: str
    day: str
    time: str
    size: int

##

import seaborn as sns

tips_df = sns.load_dataset("tips")
print(tips_df.head())

###

def orders_from_df(df) -> list[Order]:
    orders = []

    for _, row in df.iterrows():
        order = Order(
            total_bill=float(row["total_bill"]),
            tip=float(row["tip"]),
            sex=str(row["sex"]),
            smoker=str(row["smoker"]),
            day=str(row["day"]),
            time=str(row["time"]),
            size=int(row["size"]),
        )
        orders.append(order)

    return orders


orders = orders_from_df(tips_df)

print(len(orders))
print(orders[0])

####
dinner_orders = []
for order in orders:
    if    order.time == "Dinner":
        dinner_orders.append(order)
suma_chekov = 0
for order in dinner_orders:
    suma_chekov += order.total_bill
print("*"*50)
print(suma_chekov)
print(dinner_orders)

tip_percentages = []
kolichesto_chek = len(dinner_orders)
for order in dinner_orders:
    tip_percentages.append(round(order.tip / order.total_bill * 100))
print(tip_percentages)
print(kolichesto_chek)
x = (suma_chekov / kolichesto_chek)
print(x)
y = sum(tip_percentages)/ len(tip_percentages)
print(y)
```

```python
def avg_bill(orders):
    if not orders:
        return 0.0
    total_bill = sum(o.total_bill for o in orders)
    print(total_bill)
    avg_bill = total_bill / len(orders)
    print(avg_bill)
    return avg_bill

```

```python
sat_orders = [x for x in orders if x.day == "Sat"]
print(avg_bill(sat_orders))
```
