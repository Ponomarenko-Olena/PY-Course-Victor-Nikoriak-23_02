# Алгоритм обробки чеків

```python
print(len(orders))
```

```python
dict_tips_waiter = {}

i = 0
while i < len(orders):
    print(orders[i])
    order = orders[i]
    waiter = order[0]
    tip = order[2]

    if waiter not in dict_tips_waiter:
        dict_tips_waiter[waiter] = 0

    dict_tips_waiter[waiter] += tip
    i += 1

for waiter, tip in dict_tips_waiter.items():
    print(waiter, tip)

```

# Пошук лідера

```python
waiters_list = list(dict_tips_waiter.items())
print(waiters_list[0])

leader = waiters_list[0][0]
max_tips = waiters_list[0][1]

print(f"leader: {leader}, max_tips: {max_tips}")

i = 1

while i < len(waiters_list):
    waiter, tips = waiters_list[i]

    if tips > max_tips:
        leader = waiter
        max_tips = tips
        print(f"New leader: {leader}, max_tips: {max_tips}")
    else:
        print(f"{waiter} ({tips} грн) - не перевищує  {max_tips}")

    i +=1

print('='*50)
print(f'Лідер за зміну: {leader} максимальні чайові становили {max_tips} грн')



```

```python
leader_v2 = max(dict_tips_waiter, key=dict_tips_waiter.get)

print(leader)
print(leader_v2)
print(f"Однакові {leader==leader_v2}")

leader_v3, max_v3 = max(dict_tips_waiter.items(), key= lambda item: item[1])
print(f" {leader_v3} -> {max_v3}")
```

# Унікальний офіціант

```python
unique_waiters = set()

i = 0
while i < len(orders):
    waiter = orders[i][0]
    unique_waiters.add(waiter)
    i += 1

print(f" Унікальних офіціантів  ->  {unique_waiters}")
print('='*50)
print(f"Кількість                    {len(unique_waiters)}")



```

```python
unique_waiters_2 = set()

for order in orders:
    print(f'{order} - {len(orders)}order')
    start = len(unique_waiters_2)
    unique_waiters_2.add(order[0])
    finish = len(unique_waiters_2)
    status = "Додано" if finish > start else "вже є"
    print(f"[{status}] '{order[0]:<10}' -> {unique_waiters_2}")


```

```python
from collections import Counter

# Скільки чеків обслужив кожен офіціант?
orders_count = Counter(waiter for waiter, bill, tip in orders)

print("Кількість чеків по офіціантах:")
for waiter, count in orders_count.most_common():
    print(f"  {waiter:<12}: {count} чеки")
print()

# Тепер порахуємо середні чайові на чек для кожного
print("Середні чайові на один чек:")
for waiter in sorted(dict_tips_waiter):
    total = dict_tips_waiter[waiter]
    count = orders_count[waiter]
    avg   = total / count
    print(f"  {waiter:<12}: {avg:>6.1f} грн/чек  "
          f"(всього {total} грн за {count} чеки)")


```
